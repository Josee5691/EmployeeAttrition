```python
categorical_columns = data.select_dtypes(include=['object']).columns

# Setting up subplots
num_plots = len(categorical_columns)
num_cols = 2
num_rows = (num_plots + 1) // num_cols

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 4 * num_rows))

for i, column in enumerate(categorical_columns):
    row = i // num_cols
    col = i % num_cols

    # Create count plot
    sns.countplot(x=column, data=data, ax=axes[row, col])
    axes[row, col].set_title(f'Count plot for {column}')
    axes[row, col].set_xticklabels(axes[row, col].get_xticklabels(), rotation=45, horizontalalignment='right')

# If the number of plots is not a perfect multiple of num_cols, remove empty subplots
if num_plots % num_cols != 0:
    for i in range(num_plots, num_rows * num_cols):
        fig.delaxes(axes.flatten()[i])

plt.tight_layout()
plt.show()
```
